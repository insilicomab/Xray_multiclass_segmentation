### 画像の可視化

```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image

from glob import glob
```

```python
n_max_imgs = 5

plt.figure(figsize=(9, 20))
for n in range(n_max_imgs):

    # 入力用の画像データ
    image = Image.open(f'/content/segmentation02/segmentation/org_train/case00{n+1}.bmp')
    image = np.array(image)

    # 出力用の画像データ
    label = Image.open(f'/content/segmentation02/segmentation/label_train/case00{n+1}_label.png')
    label = np.array(label)

    plt.subplot(n_max_imgs, 2, 2*n+1)
    plt.imshow(image, cmap='gray')

    plt.subplot(n_max_imgs, 2, 2*n+2)
    plt.imshow(label, cmap='gray')
```
